## Duplicate Value Analysis

The dataset was checked for duplicate records using
`df.duplicated().sum()`.

No duplicate rows were found in the dataset, indicating
good data consistency and integrity.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
df['Order Year'] = df['Order Date'].dt.year

df['Order Month'] = df['Order Date'].dt.month

df['Order Day'] = df['Order Date'].dt.day
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

sns.boxplot(x=df['Sales'])

plt.title("Outlier Detection in Sales")

plt.show()
```

## Outlier Detection

Outlier analysis was performed using a boxplot on the `Sales`
column to identify extreme values and understand data distribution.

The visualization helps detect unusually high or low sales values
that may influence analytical results.

```python
plt.figure(figsize=(0,0))

corr = df.corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Matrix")

plt.show()
```
